```python
#simulation of two dices:
def dices(r,x,z,p):
    import random
    w=0
    y=0
    t=1
    counter=0
    while(t<=r):
        dice1=random.randint(w,x)
        dice2=random.randint(y,z)
        if p=="Y":
            print("For the roll no:",counter,"Dice 1 outcome is:",dice1,"Dice 2 outcome is:",dice2)
            t=t+1
            counter=counter+1
            if dice1==dice2:
                print("Sorry but you can't roll doubles!")
                break
        else:
            counter=counter+1
            print("For the roll no:",counter,"Dice 1 outcome is:",dice1,"Dice 2 outcome is:",dice2)
            t=t+1
D1=int(input("Please input the sides of the first dice:"))
D2=int(input("Please input the sides of the second dice:"))
R=int(input("Please input the no of times you would like to roll the dice:"))
P=input("Would you like to enforce the double roll policy?")
dices(R,D1,D2,P)
```

```python
#. Construct a class that can represent times. The class will have member variables hours and minutes. Equip the class with a member function
class times():
    def __init__(self,hours,minutes):
        self._hours=hours
        self._minutes=minutes
    def file_creator(self):
        self._hours=str(self._hours)
        self._minutes=str(self._minutes)
        fileopener=open("timeadder.txt","a") #appending automatically goes into the next line
        fileopener.write(self._hours+" ")
        fileopener.write(self._minutes+"\n")
        fileopener.close()
        
    def filereader(self):
        filereader=open('timeadder.txt',"r")
        lst=[]
        for line in filereader:
            line=line.strip()
            line=line.split(" ") #returns elements in a list,removes what ever stated in it
            if "" not in line: #if the line is not empty. The first line is empty, line.split removes the space between this.It becomes=""
                for el in line:
                    lst.append(int(el))
            
                
        print(lst)
        filereader.close()
        return(lst)
    
    def _add_(self,other):
        totalminutes=0
        length=len(other)
        for i in range(length):
            if i%2==0:
                totalminutes+=other[i]*60 
            else:
                totalminutes+=other[i]
        actualtime=totalminutes/60
        actualtimeminutes=totalminutes%60
        print(int(actualtime),actualtimeminutes)
#empty file
#write to the file with a new line
#line.split(" ") returns the result in a list

time1=times(3,20)
time2=times(4,15)
time3=times(6,35)
time4=times(2,25)
time5=times(0,0)
time1.file_creator()
time2.file_creator()
time3.file_creator()
time4.file_creator()
receiver=time5.filereader()
time5._add_(receiver)
```

```python
#Program of python/example-of-inheritance-with-two-child-derived-classes-in-python.
class languages:
    def __init__(self,name,OOP,stringformatting): #In inheritance, object of a parent class will have method of parent class
        self._name=name
        self._OOP=OOP
        self._stringformatting=stringformatting
    def printfeatures(self):
        print("This is a parent method.")
    
class python(languages):
    def __init__(self,portable,name,OOP,stringformatting):
        self._portable=portable
        super().__init__(name,OOP,stringformatting)
    def printfeatures(self):
        print("The name of the language is:",self._name)
        print("OOP:",self._OOP)
        print("The string formatting is:",self._stringformatting)
        print("Portable:",self._portable)
        
class java(languages):
    def __init__(self,robust,name,OOP,stringformatting):
        self._robust=robust
        super().__init__(name,OOP,stringformatting)
    def printfeatures(self):
        print("The name of the language is:",self._name)
        print("OOP:",self._OOP)
        print("The string formatting is:",self._stringformatting)
        print("Robust:",self._robust)
        
language1=python("yes","Python","Yes","C-style formatting")
language2=java("Yes","Java","Yes","C-style formatting")
language1.printfeatures()
language2.printfeatures()
languages.printfeatures(language1) #accessing a parent class with a child class object
```

```python
#Given an array, return the number of its elements that appear in it an odd number of times. For instance, on the array [0, 5, 8, -1, 5, 7, 8, 5, 7, 0, -6] returns 3 (as –1, 5, and –6 appear an odd number of times in it):

ar=[0,1,10,15,15,16,20,20,23,23,2,1,4] #count the number of occurences
def array_odd(ar):
    dict1={}
    for el in ar:
        if el not in dict1.keys(): #occurences are being stored in dictionary
            dict1[el]=1 #it is passing the key and assigning it a value
        else:
            dict1.update({el:dict1[el]+1})
    lst=[]
    for keys,values in dict1.items():
        if values%2!=0:
            lst.append(keys)
    return lst
    
t=array_odd(ar)
print(t)
```

```python
#bubblesort:
import numpy
import timeit
code="""
lst=[0,22,12,33,27,200,99,67,59,48,151,17,242,300,1]
def bubblesort(theSeq):
    n=len(theSeq)
    for i in range(n-1):                    
        for j in range(n-1):                            #it will start from the first element till the last
            if theSeq[j]>theSeq[j+1]:
                tmp= theSeq[j]
                theSeq[j]=theSeq[j+1]
                theSeq[j+1]=tmp
    return theSeq

finalresult=bubblesort(lst)
print(finalresult)"""

elapsedtime=timeit.timeit(code,number=5)/5
print(elapsedtime)      
avg=numpy.mean([0.00023578,0.00022969,0.00015765,0.00016506,0.00027502])
print("This is the average time:",avg)              
```

```python
#merging of two lists algo:
def mergelst(lstA,lstB):
    a=0
    b=0
    newlst=[]
    
    while a<len(lstA) and b<len(lstB):
        if lstA[a]<lstB[b]:
            newlst.append(lstA[a])
            a+=1
        else:
            newlst.append(lstB[b])
            b+=1
    while a<len(lstA):
        newlst.append(lstA[a])
        a+=1
    while b<len(lstB):
        newlst.append(lstB[b])
        b+=1
    return newlst

lstA=[2,4,6,9,11,22,33,44]
print(len(lstA))
lstB=[1,4,5,6,9,10,12,14,15,44,66]       #the lists need to be sorted 
print(len(lstB))

finallst=mergelst(lstA,lstB)
print("The final lst:",finallst)
```

```python
#Write a python program that ask user to enter values for two lists untill user press -1,sort these two lists using bubble sort and then merge these two lists:
def bubblesort(y):
    for i in range(len(y)):     #the values=5, depends on the values you want to sort, that is why len, not len-1
        for j in range(len(y)-1): #comparision between two values, otherwise it will go out of index
            if y[j]>y[j+1]:       #you push the greater number on the right side
                temp=y[j]
                y[j]=y[j+1]
                y[j+1]=temp
    return y

def mergelst(lst,lst2):
    newlst=[]
    a=0
    b=0
    
    while a<len(lst) and b<len(lst2):
        if lst[a]<lst2[b]:
            newlst.append(lst[a])
            a+=1
        else:
            newlst.append(lst2[b])
            b+=1
            
    while a<len(lst):
        newlst.append(lst[a])
        a+=1
    
    while b<len(lst2):
        newlst.append(lst2[b])
        b+=1
        
    return newlst

lst=[]
user=0
print("Enter -1 to exit!")
while (user!=-1):
    user=int(input("Please enter a value:"))
    if user>=0:
        lst.append(user)
print("Unsorted first list:",lst)

lst2=[]
user2=0
print("Please enter -2 to exit!")
while(user2!=-2):
    user2=int(input("Please enter a value:"))
    if user2>=0:
        lst2.append(user2)
print("Unsorted second list:",lst2)

flst=bubblesort(lst)
slst=bubblesort(lst2)
print("This is the first sorted list:",flst)
print("This is the second sorted list:",slst)

finallst=mergelst(lst,lst2)
print("This is the final list:",finallst)
```

```python
# Define a function named dif2 that accepts an integer N as input
#parameter and constructs and returns an N × N two-dimensional
#numpy array A, with the value -2.0 on the main diagonal and the
#value +1.0 on the super-diagonal and the sub-diagonal
import numpy
def dif2(N):
    global rows
    global columns
    
    rows=N
    columns=N
    lst=[]
    for r in range(rows):
        lst.append([])
        for c in range(columns):
            if r==c:
                lst[r].append(-2)
            else:
                lst[r].append(1)
    return lst

N=int(input("Please enter the integer N:"))
array=numpy.array(dif2(N))
print("This is your array:",array)
```

```python
# implementation of queue via simple list:
class Queue:
    def __init__(self):
        self._qlist=list()
    def isEmpty(self):
        return len(self._qlist)==0
    def length(self):
        return len(self._qlist)
    def enqueue(self,item):
        self._qlist.append(item)
    def dequeue(self):
        if not(self.isEmpty()):
            return self._qlist.pop(0)
        else:
            print("The Queue is empty.") #pop(normally removes the last element of the list)
    def traverse(self):
        while not(self.isEmpty()):
            print(self.dequeue())
        
q1=Queue()
userinput=0
while (userinput!=-1):
    userinput=int(input("Please enter a value:"))
    if userinput!=-1:
        q1.enqueue(userinput)

length=q1.length()
print(length)
#removal=q1.dequeue()
#print("removal:",removal)
#removal2=q1.dequeue()
#print("removal 2:",removal2)
#length=q1.length()
#print("length:",length)
q1.traverse()
empty=q1.isEmpty()
print("Empty:",empty)
removal=q1.dequeue()
print(removal) #nothing to return over here that is why it is returning None. if the queue is empty.
```

```python
#write a program that ask user to studentRecords such as studentName,StudentId,studentDegree.
#Keep on asking data from user until user wants.
#Then put the student record-such as name,id and degree as single unit in the queue.
#After that user to enter id of student to search for in queue,
#if that id is at top dequeue it othewise give  a message student is in qeue

class Queue:
    def __init__(self):
        self.qlist=list()
    
    def enqueue(self,item):
        self.qlist.append(item)
    def dequeue(self):
        return self.qlist.pop(0)
    def search(self,item):
        for el in self.qlist:
            if el.studentid!=item:
                print("This is an invalid student id!")
            else:
                break
            
        if self.qlist[0].studentid==item:
            print("ID is at the front:")
            removal=self.dequeue()
            return removal
        else:
            print("Student is in queue but not at the top!")
            
class Students:
    def __init__(self,studentid,studentname,studentdegree):
        self.studentid=studentid
        self.studentname=studentname
        self.studentdegree=studentdegree
            
qlist=Queue()
userinput=0
while (userinput!=-1):
    idinput=int(input("Please enter the student id:"))
    name=input("Please enter your name:")
    degree=input("Please enter the degree:")
    std=Students(idinput,name,degree)     #an object can be passed and its attributes can be accessed to perform various operations
    qlist.enqueue(std) 
    userinput=int(input("Do you wish to continue? Enter -1 to exit or anything else to continue."))

dequeueinput=int(input("Please enter a valid student id to search:"))
obj=qlist.search(dequeueinput) 
print("Item dequeued:",obj.studentid,obj.studentname,obj.studentdegree)# you need to pass a function with reference of its object
```
